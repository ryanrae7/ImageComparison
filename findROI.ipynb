```python
%matplotlib inline
from PIL import Image, ImageChops, ImageTk, ImageEnhance, ImageFilter
import numpy as np
import tkinter as tk
from tkinter import filedialog
import decimal
import pandas as pd
import math
import os
import cv2
import pathlib
import matplotlib.pyplot as plt

image_path = "C:/Users/rgae/Downloads/D58850_4.02.05/3400-English/Setup.JPG"

#Open the image file
image = Image.open(image_path)

#convert the data of the image --> array
data = np.array(image)

#Show image using mathplotlib
plt.imshow(data, interpolation='nearest')
plt.axis('off') 
plt.show()

#Print array of the dimension
print(f"Image shape: {data.shape}")

#Input the coordinates
#Coordinates are in the format (start column, end column, start row, end row, )
roi_coords = (900, 1024, 0, 50)  # Adjust these coordinates as needed

#get the roi and slice using the coordinates
roi = data[roi_coords[2]:roi_coords[3], roi_coords[0]:roi_coords[1]]

#Display the ROI using matplotlib
plt.imshow(roi, interpolation='nearest')
plt.axis('off')  #Hide the axis
plt.title('Region of Interest (ROI)')
plt.show()

#
```
